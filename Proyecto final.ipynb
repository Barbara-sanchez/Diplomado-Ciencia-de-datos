```python
#Función limpieza
def clean_text(text, pattern="[^a-zA-Z0-9]"):
    cleaned_text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
    
    cleaned_text = re.sub(pattern, " ", cleaned_text.decode("ascii"), flags=re.UNICODE)
    cleaned_text = u' '.join(cleaned_text.lower().lstrip().strip().split())
    return cleaned_text
```

```python
# "c_"  - Variables numericas : Discretas y continuas
# "v_"  - Variables categoricas
# "d_"  - Variables tipo fecha
# "t_"  - Variables de texto : comentarios, descripciones, url
d_Feats=["Date","Time","YOM"]
c_Feats=["Crew on board","Crew fatalities","Pax on board","PAX fatalities","Other fatalities","Total fatalities"]
v_Feats=["Aircraft","Operator","Flight phase","Flight type","Survivors","Crash site","Country","Region","Crash cause"]
t_Feats=["Schedule","MSN","Flight no.","Circumstances","Registration"]
```

```python
c_Feats_new=["c_"+x for x in c_Feats]
v_Feats_new=["v_"+x for x in v_Feats]
d_Feats_new=["d_"+x for x in d_Feats]
t_Feats_new=["t_"+x for x in t_Feats]
```

```python
DataNort.rename(columns=dict(zip(d_Feats,d_Feats_new)),inplace=True)
DataNort.rename(columns=dict(zip(v_Feats,v_Feats_new)),inplace=True)
DataNort.rename(columns=dict(zip(t_Feats,t_Feats_new)),inplace=True)
DataNort.rename(columns=dict(zip(c_Feats,c_Feats_new)),inplace=True)
```
